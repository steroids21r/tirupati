```python
import pandas as od
path="C:\\Users\\Kumar R\\OneDrive\\Documents\\Downloads\\Coimbatore.csv"
data=od.read_csv(path)
print(data)
print(data.info())

data = data.fillna(method='ffill')


print(data['State_Name'].unique())
print(data['State_Name'].nunique())
data['State_Name']=data['State_Name'].map({'Tamil Nadu':1})
print(data['ind_district'].unique())
print(data['ind_district'].nunique())
data['ind_district']=data['ind_district'].map({'coimbatore':0})
#print(data)

inputs=data.drop('X4','columns')
output=data['X4']
print(inputs)
print(output)

import sklearn
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(inputs,output)

#y=m1*x1+m2*x2*+m3*x3+b

State_Name=int(input("enter the State_Name"))
ind_district=int(input("enter the ind_district "))
x2=int(input("enter the x2"))
y=int(input("enter the y"))
x1=int(input("enter the x1"))
x3=int(input("enter the x3"))
print(model.predict([[State_Name,ind_district,x2,y,x1,x3]]))

coef=model.coef_
print(coef)

intercept=model.intercept_
print(intercept)

you=coef*15034+intercept
print(you)
```
